## 앞으로 확인해야할 점
1. 한 wafer가 들어가서 나오기까지의 데이터를 살펴보면서, 한 wafer내에서의 컬럼들 변화를 살펴보기

```python
def extract_a_wafer(df):
    wafer_list = list()
    wafers = df['Lot'].values

    start=0
    for idx in range(len(wafers)):

        if idx+1 > len(wafers)-1:
            wafer_list.append(df.iloc[start:idx+1])
            return wafer_list

        if wafers[idx] != wafers[idx+1]:
            wafer_list.append(df.iloc[start:idx+1])
            start = idx+1

wafer_list = extract_a_wafer(df)
```

```python
wafer_list[4].describe()
```

```python
tmp = list()
for wafer in wafer_list:
    if np.std(wafer['runnum']) != 0:
        print(wafer['Lot'])
    tmp.append([np.std(wafer['Lot']),np.std(wafer['stage']),np.std(wafer['runnum']),np.std(wafer['recipe']),np.std(wafer['recipe_step'])])
```

```python
tmp_df = pd.DataFrame(tmp, columns=['Lot', 'stage','runnum','recipe','recipe_step'])
tmp_df.describe()
```

## 발견점
1. 하나의 Wafer가 공정에 들어가면, 하나의 processing stage 거친다.
2. 하나의 Wafer 공정당, 한개의 Recipe에 해당하는 공정을 거친다.
3. 1과 2를 종합하면, 하나의 Wafer 공정당 한개의 Recipe대로 공정이 이루어지고 이때의 processing stage는 하나이다.
4. 대부분의 경우에는 하나의 Wafer가 공정 중일 때 runnum은 변하지 않지만 2570,4409,8586번 Wafer는 중간에 runnum의 변화가 있었음
